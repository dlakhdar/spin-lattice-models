```python
import numpy as np 
import matplotlib.pyplot as plt 

plt.style.use("bmh")# mappin functions 

x = lambda u, v : u + np.cos(v)*np.exp(u)
y = lambda u, v : v + np.sin(v)*np.exp(u)

#w-plane initial
fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns

# Draw a horizontal line at y = 2
axes[0].axhline(y=np.pi, color='blue', linestyle='--', linewidth=2, label="y = π")
axes[0].axhline(y=-np.pi, color='red', linestyle='--', linewidth=2, label="y = -π")

axes[0].set_xlim(-5, 5)  # Set x-axis limits
axes[0].set_ylim(-np.pi-2, np.pi+2)  # Set y-axis limits

axes[0].axhline(y=0, color="black", linewidth=1)  # X-axis
axes[0].axvline(x=0, color="black", linewidth=1)  # Y-axis
# Set Cartesian grid
axes[0].grid(visible=True, which="both", linestyle="--", linewidth=0.5, color="gray")
axes[0].set_xlabel("u")
axes[0].set_ylabel("v")
axes[0].set_title("w-plane")

y_consts= np.arange(-2,3,1)
for const in y_consts:
    axes[0].axhline(y=const, color='black', linestyle='--', linewidth=2)


# Manually create legend handles
legend_handles = [
    plt.Line2D([0], [0], color="black", label="v-lines"),
]

axes[0].legend(handles=legend_handles, loc="upper right")

# Set equal aspect ratio to ensure square grid spacing


# Labels and title
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].set_title("z-plane")

# Draw a horizontal line at y = 2
axes[1].axhline(y=np.pi, xmin=0,xmax=.4, color='blue', linestyle='--', linewidth=2, label="y = π")
axes[1].axhline(y=-np.pi, xmin=0,xmax=.4, color='red', linestyle='--', linewidth=2, label="y = -π")

axes[1].set_xlim(-15, 15)  # Set x-axis limits
axes[1].set_ylim(-np.pi-6, np.pi+6)  # Set y-axis limits

axes[1].axhline(y=0, color="black", linewidth=1)  # X-axis
axes[1].axvline(x=0, color="black", linewidth=1)  # Y-axis
# Set Cartesian grid
axes[1].grid(visible=True, which="both", linestyle="--", linewidth=0.5, color="gray")

# plot equipotential lines 

u , v = np.arange(-np.pi,np.pi,.5),np.arange(-np.pi,np.pi,.5)

for const in v:
    axes[1].plot(x(u,const),y(u,const),color="black",linewidth=.75)

for const in u:
    axes[1].plot(x(const,v),y(const,v),color="grey",linewidth=.5)


# Manually create legend handles
legend_handles = [
    plt.Line2D([0], [0], color="black", label="E-lines"),
    plt.Line2D([0], [0], color="grey", label="Equipotential-Lines")
]

# Add the legend outside the loop
axes[1].legend(handles=legend_handles, loc="upper right",fontsize='xx-small')

# Show the plot
plt.show()
```
