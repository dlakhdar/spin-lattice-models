```python
import numpy as np 
import matplotlib.pyplot as plt 


ρ = 1
u = lambda ρ, x,y: ρ*(1-(x**2+y**2))/(x**2+y**2+2*y+1)
v = lambda ρ, x,y: 2*ρ*x/(x**2+y**2+2*y+1)
```

```python


C = 2.0
# line constant x
y = np.arange(0.0,2.0,.1) 
x = np.zeros(len(y))
x.fill(C)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns

# Left plot: z-plane
axes[0].set_title("z-plane")
axes[0].plot(x, y, label="Original (x, y)")
axes[0].grid()
axes[0].legend()

# Right plot: w-plane
colors = ["blue", "red", "green", "orange", "purple", "cyan", "magenta", "yellow"]
axes[1].set_title("w-plane")
for c,n in zip(colors, range(2,len(colors)+2)):
    axes[1].plot(u(n, x, y), v(n, x, y), color="orange", label="Transformed (u, v)")
    
axes[1].grid()
axes[1].legend()

# Show the side-by-side plots
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
X = lambda ρ, u, v : (2*v*ρ)/(u**2+v**2+2*u*ρ+ρ**2)
Y = lambda  ρ, u, v : -1*(u**2+v**2-ρ)/(u**2+v**2+2*u*ρ+ρ**2)

# Define circle parameters
radius = 1  # Adjust radius
theta = np.linspace(0,  np.pi, 300)  # Angle range from 0 to 2π

# Parametric equations for a circle
u = radius * np.cos(theta)
v = radius * np.sin(theta)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns

# Left plot: z-plane
axes[0].set_title("w-plane")
axes[0].plot(u, v, label="Original (u, v): unit circle")
axes[0].grid()
axes[0].legend()

# Right plot: w-plane
colors = ["blue", "red", "green", "orange", "purple", "cyan", "magenta", "yellow"]
for c,n in zip(colors, range(2,len(colors)+2)):
    axes[1].set_title("z-plane")
    axes[1].plot(X(n, u, v), 
                Y(n, u, v), 
                color=c, 
                label=f"Transformed (x, y): ρ={n}")

axes[1].grid()
axes[1].legend()

# Show the side-by-side plots
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
